```python
import os
import urllib.request

# to read excel
#df = pd.read_excel(file_https://github.com/fenago/MLEssentials/blob/main/datasets/Online%20Retail.xlsx?raw=trueurl)
### data_location = 'https://raw.githubusercontent.com/elephantscale/datasets/master/retail-data/all/online-retail-dataset.csv'

# sample data - 25k rows
data_location = '../data/online-retail-sample.csv'
data_url = 'https://raw.githubusercontent.com/elephantscale/datasets/master/retail-data/all/online-retail-sample.csv'

## large dataset here (500k+ rows) -- uncomment this line
# data_url = 'https://github.com/elephantscale/datasets/blob/master/retail-data/all/online-retail-dataset.csv?raw=true'


if not os.path.exists(data_location):
    data_location = os.path.basename(data_location)
    if not os.path.exists(data_location):
        print ("Downloading  ", data_url)
        urllib.request.urlretrieve(data_url, data_location)
print ('data_location: ', data_location)
```

```python
import pandas as pd

df = pd.read_csv(data_location)
df
```

```python
# see data types
# we see 'invoicedate' is object
df.dtypes
```

## Parse Dates

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.dtypes
```
